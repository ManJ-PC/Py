## FFill by Groups

```python
fruit_df = pd.DataFrame(data=[
                             [dt.datetime(2020,1,1),'apple',1,],
                             [dt.datetime(2020,1,2),'apple',2],
                             [dt.datetime(2020,1,3),'apple',None],
                             [dt.datetime(2020,1,4),'apple',None],
                             [dt.datetime(2020,1,5),'apple',5],
                             [dt.datetime(2020,1,1),'orange',4,],
                             [dt.datetime(2020,1,2),'orange',50],
                             [dt.datetime(2020,1,3),'orange',None],
                             [dt.datetime(2020,1,4),'orange',55],
                             [dt.datetime(2020,1,5),'orange',None],
                              [dt.datetime(2020,1,1),'banana',None],
                              [dt.datetime(2020,1,2),'banana',None],
                              [dt.datetime(2020,1,3),'banana',None],
                              [dt.datetime(2020,1,4),'banana',None],
                              [dt.datetime(2020,1,5),'banana',None],
                             ],
                       columns=['date','fruit','quantity'])
fruit_df
```

```python
#A straight up ffill will run into categories that are all blank

fruit_df.groupby(['fruit',pd.Grouper(key='date',freq='D')])['quantity'].mean().ffill().reset_index()
```

```python
#Using a Lambda function, we can perform the operation by grouping and adding the filled column
fruit_df["quantity"] = fruit_df.groupby('fruit')['quantity'].transform(lambda x: x.ffill())
fruit_df
```

```python
#Using a for loop, we can loop through each grouping. This is more expensive
fruit_df.set_index(['fruit','date'], inplace=True)

lst = fruit_df.index.get_level_values(level=0).unique()
```

```python
df_new = pd.DataFrame()

for i in lst:
  df_new = pd.concat([df_new,fruit_df.loc[i].ffill()])

df_new
```

##How to Dynamically Format Pandas DataFrame Columns to be Database and Parquet Ready

This is helpful for making sure that your columns are SQL ready.

https://www.pauldesalvo.com/how-to-dynamically-format-pandas-dataframe-columns-to-be-database-and-parquet-ready/

```python
df = pd.DataFrame(np.random.randint(100,size=(5, 12)),columns=['Column with @#$@#$ characters','Question #1','Test?','Test','TEST','Hello','hello', 'test.testing', 'comma,testing',0,1,2])
df
```

```python
df.to_parquet('test.parquet')
```

```python
df.columns = (df.columns
              .astype(str) #convert all the elements to a string because Pandas allows you to have numbers for column names
              .str.replace('[.,]', '_') #Replace specific symbols with underscores
              .str.extractall(r'([\w\d]+)') #extract just words and numbers
              .unstack() #bring all the groups back into one row
              .apply(lambda x: '_'.join(x.dropna() #replace empty space with an underscore
              .astype(str)),axis=1) #convert column names to a string
              .str.lower() #lowercase
              .str.strip() #clean up white spaces if applicable
              )

df
```

```python
#Check for duplicate columns
df.columns.duplicated().any()
```

```python
#This package takes a number and spells it out. So if you pass in 1 it would return 'one'
!pip install inflect
import inflect
p = inflect.engine()
```

```python
def create_unique_cols(x):
  col_lst = [] #existing col list
  unique_col_lst = [] #new col list with modified col names for the duplicates

  for i in x.columns:
    col_lst.append(i) #build the list as you go through the elements

    if i.isdecimal(): #convert numbers to a string. In my case, I am just spelling out the number
      i = p.number_to_words(i)
    else:
      pass

    if col_lst.count(i) > 1:
      unique_col_lst.append(f'{i}_{col_lst.count(i)}') #add the duplicated column with a number
    else:
      unique_col_lst.append(i) #add the column as is to the new list

  return unique_col_lst
```

```python
df.columns = create_unique_cols(df)
```

```python
df
```

```python
#Now running to parquet will work
df.to_parquet('test.parquet')
```

## Compare Columns in Two DataFrames

```python
df1 = pd.DataFrame([[1,2,3],[1,2,3],[1,2,3]],columns=['A','B','C'])

df2 = pd.DataFrame([[1,2,4],[1,2,4],[1,2,4]],columns=['A','B','D'])
```

```python
df2
```

```python
def compare_columns(x,y):
  columns1 = x.columns.tolist()
  columns2 = y.columns.tolist()

  same_columns = list(set(columns1).intersection(columns2))
  columns_not_in_first = list(set(columns2) - set(columns1))
  columns_not_in_second = list(set(columns1) - set(columns2))

  result = f"same colums: {same_columns}\ncolumns not in first dataframe: {columns_not_in_first}\ncolumns not in second dataframe: {columns_not_in_second}"

  return result
```

```python
print(compare_columns(df1,df2))
```

## Alternatives to using Pandas Apply to speed up calculated column performance

```python
df = pd.DataFrame(
    np.random.randint(100,size=(1_000_000, 2))
    ,columns=['A','B'])
df
```

```python
time1 = dt.datetime.now()

df.apply(lambda x: x['A'] + x['B'], axis=1)

time2 = dt.datetime.now()

print(f"Function took {time2-time1} time to run.")
```

```python
time1 = dt.datetime.now()

df['A'] + df['B']

time2 = dt.datetime.now()

print(f"Function took {time2-time1} time to run.")
```

```python
#conditional using pandas apply
time1 = dt.datetime.now()

df['C'] = df.apply(lambda x: 'under 50' if x['A'] <= 50 else 'over 50', axis=1)

time2 = dt.datetime.now()

print(f"Function took {time2-time1} time to run.")
print(df)
```

```python
#conditional using numpy
time1 = dt.datetime.now()

df['C'] = np.where(df['A'] <= 50, 'under 50', 'over 50')

time2 = dt.datetime.now()

print(f"Function took {time2-time1} time to run.\n")
print(df)
```

## Chart Emojis using Plotly

```python
!pip install -U pip setuptools wheel
!pip install -U spacy
!python -m spacy download en_core_web_sm
!pip install spacymoji

import spacy
from spacymoji import Emoji

nlp = spacy.load("en_core_web_sm")
emoji = Emoji(nlp)
nlp.add_pipe("emoji", first=True)

nlp.pipe_names
```

```python
text = ['Hey check this out 👍👍👍👍👍👍',
        'Happy Birthday 🎉🎉🎉🎉🎊🎁',
        'That is so funny 😂😂🤣🤣🤣',
        '👏👏👏']
```

```python
#Convert text list to a Pandas DataFrame
df = pd.DataFrame(text, columns=['text'])
df
```

```python
def extract_emojies(x):
  doc = nlp(x['text']) #with emojis
  emojis = [token.text for token in doc if token._.is_emoji]

  return emojis
```

```python
emojies_df = df.apply(extract_emojies,axis=1)
emojies_df
```

```python
emoji_counts = (emojies_df
                .apply(pd.Series)
                .stack()
                .value_counts()
                .rename('Count')
                .sort_values()
                .reset_index()
                .rename(columns={'index':'Emoji'}))
emoji_counts
```

```python
import plotly.graph_objects as go
import plotly.io as pio
import plotly.express as px
import plotly

fig = go.Figure()

fig.add_trace(go.Scatter(y=emoji_counts['Emoji'],
                x=emoji_counts['Count'],
                name='Emoji Counts',
                marker_color='white',
                orientation='h',
                text=emoji_counts['Emoji'],
                textposition='top center',
                mode='markers+text',
                textfont=dict(size=30),
                ))

fig.update_yaxes(visible =False, range=[0, len(emoji_counts)+1])
fig.update_xaxes(title='Number of Times Used')

fig.update_layout(
    template='simple_white',
    height=len(emoji_counts)*75, width = 800)
```

# Reverse Address Lookup

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import pandas as pd

locator = Nominatim(user_agent="my-application")

def get_adress(x):
  '''
  Takes in latitude and longitude and returns and address
  '''
  coordinates = [x.latitude,x.longitude]
  location = locator.reverse(coordinates)
  address = location.raw
  address_converted = pd.json_normalize(address).squeeze() #squeeze converts a dataframe to a pandas series

  return address_converted
```

# Get Country and State from a Phone Number

```python
!pip install phonenumbers pycountry us
import phonenumbers
from phonenumbers import carrier, timezone, geocoder
from phonenumbers.phonenumberutil import (region_code_for_country_code,region_code_for_number)
import pycountry
import us

def phone_number_info(x):
  try:
    number = phonenumbers.parse("+"+str(x), None) #first parse the phone number

    valid_phone_number = phonenumbers.is_valid_number(number) #check if valid
    number_country_code = number.country_code #grab the country code
    number_timezone = timezone.time_zones_for_number(number)[0] #timezone
    try:
      carrier = carrier.name_for_number(number, 'en')
    except:
      carrier = None

    number_region = geocoder.description_for_number(number, 'en') #Phone number Region usually the country
    country_name = pycountry.countries.get(alpha_2=region_code_for_number(number)).name
    alpha_2_code = pycountry.countries.get(alpha_2=region_code_for_number(number)).alpha_2
    alpha_3_code = pycountry.countries.get(alpha_2=region_code_for_number(number)).alpha_3

    #For US or canada numbers, get the area code so we can get more specific on the location
    if (number_country_code == 1) & (alpha_3_code == 'USA'):
      us_area_code = int(str(x)[1:4])
      state_abbr = us.states.lookup(number_region).abbr
    else:
      us_area_code = None
      state_abbr = None

  except:
    number = None
    valid_phone_number = None
    number_country_code = None
    number_timezone = None
    carrier = None
    number_region = None
    country_name = None
    alpha_2_code = None
    alpha_3_code = None
    us_area_code = None
    state_abbr = None

  return valid_phone_number, number_country_code, number_timezone, carrier, number_region, country_name, alpha_2_code, alpha_3_code, us_area_code, state_abbr
```
